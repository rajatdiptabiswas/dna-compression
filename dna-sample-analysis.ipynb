# DNA FASTA Protein Analysis

#### Importing `Biopython` library 

```python
from Bio import SeqIO
```

#### Getting `sample.fa` into a list

```python
fasta_sequences = SeqIO.parse(open('data/sample.fa'),'fasta')
seq = [str(fasta.seq) for fasta in fasta_sequences]

# printing last 5 sequences
print(seq[-5:])
```

#### Creating single string of sequence data

```python
seq_str = ''

for s in seq:
    seq_str += s
```

```python
# showing first 1000 bases of sequence
seq_str[:1000]
```

```python
# printing length of string sequence
len(seq_str)
```

Length of created string is **2294810 characters**

#### Get size of original string

```python
import sys

## size in bytes
str_size_b = sys.getsizeof(seq_str)

## size in megabytes
str_size_mb = str_size_b / 1024

str_size_mb
```

The size of the string in memory is **2241.07 MB**

#### Unique characters in the string

```python
print(set(seq_str))

# number of unique characters
print(len(set(seq_str)))
```

There are **22** unique characters in the string

#### Counting occurences of all the bases in the sequence

```python
from collections import Counter 

# using collections.Counter() to get count
char_count = Counter(seq_str) 
  
print(char_count)
```

```python
char_count.most_common()
```

#### Histogram representation for base frequency 

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
labels, values = zip(*char_count.items())
indexes = np.arange(len(labels))
```

```python
# Initialize the matplotlib figure
f1, ax1 = plt.subplots(figsize=(15, 5))

sns.barplot(x=list(labels), y=values, label="Frequency", palette="rocket")

# Add a legend and informative axis label
ax1.legend(ncol=1, loc="upper right", frameon=True)
ax1.set(xlabel="Nucleobases",
       title="Frequencies of nucleobases in the sequence string")
sns.despine(left=True, bottom=True)
```

```python
labels_h, values_h = zip(*char_count.most_common())

# reversing the lists for graphing
labels_h = list(labels_h)[::-1]
values_h = list(values_h)[::-1]
```

```python
f2, ax2 = plt.subplots(figsize=(15, 10))

plt.barh(indexes, values_h)
plt.ylabel('Bases')
plt.xlabel('Frequency')
plt.title('Frequency of bases')
plt.yticks(indexes, labels_h)
plt.show()
```
